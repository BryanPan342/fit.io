```python
import numpy as np
import os
import time

import PIL.Image as Image
import matplotlib.pylab as plt

import tensorflow as tf
import tensorflow_hub as hub

from tensorflow.keras import initializers

# Load compressed models from tensorflow_hub
os.environ['TFHUB_MODEL_LOAD_FORMAT'] = 'COMPRESSED'
```

```python
classifier_model ="https://tfhub.dev/google/imagenet/resnet_v2_50/classification/4"
```

```python
i = tf.keras.layers.Input((224, 224, 3))
core = tf.keras.applications.ResNet50V2(include_top=False, weights='imagenet')(i)
flattened = tf.keras.layers.Flatten()(core)
dense = tf.keras.layers.Dense(100, activation=tf.nn.relu, kernel_initializer=initializers.RandomNormal(stddev=0.01))(flattened)
model = tf.keras.Model(inputs=i, outputs=dense)
```

```python
model.summary()
```

```python
image = "./rashford.jpg"
img = Image.open(image)
img = img.resize((224, 224))
img = img.convert('RGB')
```

```python
plt.imshow(img)
```

```python
test_image =  np.array(img)/255.0
```

```python
test_image = np.expand_dims(test_image, axis=0)
print(test_image.shape)
```

```python
out = model([test_image])
out = np.reshape(out, (100))
print(out.shape)
```

```python
image2 = "./kai.jpg"
img2 = Image.open(image2)
img2 = img2.resize((224, 224))
img2 = img2.convert('RGB')
img2 = np.expand_dims(np.array(img2)/255.0, axis=0)
```

```python
out2 = model([img2])
out2 = np.reshape(out2, (100))
print(out2.shape)
```

```python
from scipy import spatial

marcus_kai_diff = 1 - spatial.distance.cosine(out, out2)
```

```python
print(marcus_kai_diff)
```
